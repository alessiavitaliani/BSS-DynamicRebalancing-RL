```python
import os
import sys
import subprocess
from subprocess import STDOUT, PIPE

base_path = r"C:\Users\Alessia Vitaliani\Downloads\Research training\BSS-DynamicRebalancing-RL\src\bss_rebalancing"
preproc_src = os.path.join(base_path, "preprocessing", "src")
gym_src = os.path.join(base_path, "gymnasium_env", "src")
script_file = os.path.join(preproc_src, "preprocessing", "__main__.py")

env = os.environ.copy()
env["PYTHONIOENCODING"] = "utf-8" 

paths = [gym_src, preproc_src, base_path]
env["PYTHONPATH"] = os.pathsep.join(paths) + os.pathsep + env.get("PYTHONPATH", "")

cmd = [
    sys.executable, 
    script_file, 
    "--data-path", os.path.join(base_path, "data_manhattan"),
    "--year", "2024",
    "--months", "9, 10",
    "--verbose"
]

print("🚀 Avvio Pipeline")

process = subprocess.Popen(cmd, env=env, stdout=PIPE, stderr=STDOUT, text=True, encoding='utf-8')

if process.stdout:
    for line in process.stdout:
        print(line, end="")
```
