```python
import os
import sys
import subprocess
from subprocess import STDOUT, PIPE

base_path = r"C:\Users\Alessia Vitaliani\Downloads\Research training\Bike_sharing"
preproc_src = os.path.join(base_path, "preprocessing", "src")
gym_src = os.path.join(base_path, "gymnasium_env", "src")
script_file = os.path.join(preproc_src, "preprocessing", "__main__.py")

env = os.environ.copy()
env["PYTHONIOENCODING"] = "utf-8" 

paths = [gym_src, preproc_src, base_path]
env["PYTHONPATH"] = os.pathsep.join(paths) + os.pathsep + env.get("PYTHONPATH", "")

cmd = [
    sys.executable, 
    script_file, 
    "--data-path", os.path.join(base_path, "data"),
    "--year", "2022",
    "--months", "9, 10",
    "--verbose"
]

print("🚀 Avvio Pipeline")

process = subprocess.Popen(cmd, env=env, stdout=PIPE, stderr=STDOUT, text=True, encoding='utf-8')

if process.stdout:
    for line in process.stdout:
        print(line, end="")
```

```python
# Lancio del Training 
print("Training con DQN")

# Codice Python da eseguire come stringa
python_code = f"""
import sys
import os
sys.path.extend({paths})

import gymnasium
from gymnasium.envs.registration import register

# Registrazione manuale forzata
try:
    register(
        id='gymnasium_env/FullyDynamicEnv-v0',
        entry_point='gymnasium_env.envs:FullyDynamicEnv',
    )
    print('✅ Namespace gymnasium_env registrato correttamente.')
except Exception as e:
    print(f'ℹ️ Nota sulla registrazione: {{e}}')

# Avvio del tuo script originale
import rl_training.train as train
if __name__ == "__main__":
    train.main()
"""

# Scriviamo il codice in un file temporaneo per evitare problemi di virgolette su Windows
with open("launcher_temp.py", "w", encoding="utf-8") as f:
    f.write(python_code)

# Lancio effettivo
cmd = [
    sys.executable, 
    "launcher_temp.py",
    "--data-path", os.path.join(base_path, "data"),
    "--results-path", os.path.join(base_path, "results"),
    "--num-episodes", "10",
    "--log"
]

process = subprocess.Popen(cmd, env=env, stdout=PIPE, stderr=STDOUT, text=True, encoding='utf-8')

if process.stdout:
    for line in process.stdout:
        print(line, end="")
```
